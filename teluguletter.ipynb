```
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm
```

```
DATADIR = "drive/My Drive/telugu"
CATEGORIES = ['aa','aaa','tha','ey','oo','ma']
```

```
for category in CATEGORIES:

  class_num = CATEGORIES.index(category)
  print(class_num)
```

```
import cv2
training_data = []
IMG_SIZE=50
for img,category  in  zip(tqdm(os.listdir(DATADIR)),CATEGORIES):

       img_array = cv2.imread(os.path.join(DATADIR,img),cv2.IMREAD_GRAYSCALE) # convert to array
       new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) #
       training_data.append([new_array, CATEGORIES.index(category) ]) # add this to our
```

```
training_data
```

```
x=[]
y=[]
for data in training_data:
  x.append(data[0])
  y.append(data[1])
```

```
x=np.array(x)
x.shape
```

```
x=x.reshape(6,50,25,2)
```

```
x.shape
```

```
print('The label is:', y[0])
```

```

#from keras.utils import to_categorical
y_one_hot = tf.keras.utils.to_categorical(y)

#tf.keras.utils.to_categorical(
 #   x, num_classes=None#
#)
```

```
y_one_hot
```

```
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
model = Sequential()
```

```
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
model = Sequential()
model.add(Conv2D(32, (5, 5), activation='relu', input_shape=(50,25,2)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
#model.add(Conv2D(128#, (5, 5), activation='relu'))
#model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(500, activation='relu'))
model.add(Dense(6, activation='softmax'))
```

```
#Compile the model. Give it the categorical_crossentropy loss function which is used for classes greater than 2, the adam optimizer, and the accuracy of the model.
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```
hist = model.fit(x, y_one_hot,
           batch_size=500, epochs=100, validation_split=0.3 )
```
