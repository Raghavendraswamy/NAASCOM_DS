Loads a BlenderBot model from Hugging Face.
Uses predefined FAQs for common questions (for faster responses).
If the question is not in FAQs, it generates an AI-based response.
Keeps running until the user types "exit".

```python
from transformers import BlenderbotTokenizer,BlenderbotForConditionalGeneration
# Load pre-trained model and tokenizer
model_name = "facebook/blenderbot-400M-distill"
tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)
```

```python
faq = {
    "admission requirements": "To apply, you need a high school diploma, transcripts, and SAT/ACT scores. Some programs may require additional documents.",
    "application deadline": "The deadline for applications is March 15th for Fall intake and October 15th for Spring intake.",
    "tuition fees": "The tuition fee is $10,000 per semester for in-state students and $20,000 for out-of-state students.",
    "scholarships": "We offer merit-based and need-based scholarships. You can apply through our financial aid portal.",
    "contact admissions": "You can contact the admissions office via email at admissions@college.edu or call +1-234-567-890."
}
```

```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration

# Load pre-trained model and tokenizer
model_name = "facebook/blenderbot-400M-distill"
tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)


def college_admission_chat():
    print("Welcome to the College Admission Chatbot! ")

    while True:
        user_input = input("You: ").lower()

        if user_input == "exit":
            print("Chatbot: Thank you for reaching out! Have a great day!")
            break

        # Check if the question is in FAQs
        response = None
        for key in faq:
            if key in user_input:
                response = faq[key]
                break

        if response is None:
            # Generate response using AI model
            inputs = tokenizer(user_input, return_tensors="pt")
            reply_ids = model.generate(**inputs)
            response = tokenizer.decode(reply_ids[0], skip_special_tokens=True)

        print(f"Chatbot: {response}")

# Start the chatbot
college_admission_chat()
```
