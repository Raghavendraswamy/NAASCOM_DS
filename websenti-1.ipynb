```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.trustpilot.com/review/www.tcs.com'

# Send a GET request to fetch the raw HTML content
response = requests.get(url)

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')



data=soup.find('script')
data
```

```python
script_tag = soup.find('script', type='application/ld+json')
script_tag
```

```python
 import json
 json_data = json.loads(script_tag.string)
```

```python
review_data = json_data['@graph']
review_data
```

```python
for r in review_data:
    if r.get('@type')=='Review':
        list1=[]
        list1.append(r.get('author').get('name'))
        list1.append(r.get('reviewBody'))
        list1.append(r.get('reviewRating').get('ratingValue'))
```

```python
list1[1]
```

```python
from transformers import pipeline
```

```python
classifier=pipeline('sentiment-analysis')
classifier(list1[1])
```
